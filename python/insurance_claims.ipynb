```python
import matplotlib.pyplot as plt
import seaborn as sns

fuel_claim = (df.groupby('fuel_type')['claim_status'].mean() * 100).round(2).sort_values(ascending=False)

fuel_claim.plot(kind='bar', figsize=(8,5))
plt.title('Claim Rate by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Claim Rate (%)')
plt.xticks(rotation = 0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

Avg_age = df.groupby('claim_status')['customer_age'].mean().round(2)

Avg_age.index = ['No Claim', 'Claimed']

Avg_age.plot(kind='barh')
plt.title('Claim Rate by Avearge Age')
plt.xlabel('Average Customer Age')
plt.ylabel('Claim Rate (%)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

transmission_claims = (df.groupby('transmission_type')['claim_status'].mean() * 100).round(2).sort_values(ascending=False)

transmission_claims.plot(kind='barh')
plt.title('Claim Rate by Transmission Type')
plt.xlabel('Claim Rate (%)')
plt.ylabel('Transmission Type')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

vehicle_age_claims = (df.groupby('vehicle_age')['claim_status'].mean()* 100).round(2).sort_index(ascending=True)

vehicle_age_claims.plot(kind='line', figsize=(15, 5), marker='o')
plt.title('Claim Rate by Vehicle Age')
plt.xlabel('Vechicle Age')
plt.ylabel('Claim Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

region_claim = (df.groupby('region_code')['claim_status'].mean() * 100).round(2).sort_values(ascending=False)

region_claim.plot(kind='barh', figsize=(10,15))
plt.title('Claims by Region')
plt.xlabel('Claim Rate (%)')
plt.ylabel('Region Code')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

segment_claims = (df.groupby('segment')['claim_status'].mean() * 100).round(2).sort_values(ascending=False)

segment_claims.plot(kind='barh')
plt.title('Claim Rate by Segment')
plt.xlabel('Claim Rate (%)')
plt.ylabel('Segments')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

airbags_claims = (df.groupby('airbags')['claim_status'].mean() * 100).round(2).sort_index(ascending=False)


airbags_claims.plot(kind='bar')
plt.title('Claim Rate by Airbags')
plt.xlabel('Airbags')
plt.ylabel('Claim Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ncap_claims = (df.groupby('ncap_rating')['claim_status'].mean() * 100).round(2).sort_values(ascending=False)

ncap_claims.plot(kind='bar')
plt.title('Claims by Ncap Rating')
plt.xlabel('Ncap Rating')
plt.ylabel('Claim Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
